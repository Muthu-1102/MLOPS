#**Support Vector Machine Using Digits Dataset**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
dir(digits)
```

```python
digits.data
```

```python
digits.feature_names[:5]
```

```python
for i in range(4,10):
  plt.matshow(digits.images[i])
```

```python
digits.target[:5]
```

```python
digits.target_names[:5]
```

```python
x = digits.data
y = digits.target
```

```python
from sklearn import svm
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
model = svm.SVC()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

**Accuracy of model using differnet kernals**

**Kernals - linear**

```python
model = svm.SVC(kernel='linear')
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

**Kernals - rbf**

```python
model = svm.SVC(kernel='rbf')
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

**Tunning model by changing regularation and gamma parameter**

```python
model = svm.SVC()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model = svm.SVC(gamma=10)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model = svm.SVC(C=5)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model = svm.SVC(C=10)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
model = svm.SVC()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```
