#**K Means Clustering For My Problem Statement**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("complaints_dataset.csv")
df.head()
```

```python
X = df[["Complaint_Length", "Traffic_Level"]]
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
```

```python
df["Cluster"] = kmeans.fit_predict(X)
```

```python
plt.scatter(df["Complaint_Length"],df["Traffic_Level"],c=df["Cluster"])
plt.xlabel("Complaint Length")
plt.ylabel("Traffic Level")
plt.title("KMeans Clustering")
plt.show()
```
