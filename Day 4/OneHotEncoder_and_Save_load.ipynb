```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn import linear_model
```

```python
df = pd.read_csv("dataset.csv")
df.head()
```

```python
priority_map = {"Low":1,"Medium":2,"High":3}
df["Priority"] = df["Priority"].map(priority_map)
df.head()
```

```python
dummies = pd.get_dummies(df["Category"])
dummies.head()
```

```python
merged = pd.concat([df,dummies],axis="columns")
merged.head()
```

```python
final = merged.drop(columns=["Category"])
final.head()
```

```python
x = final.drop(columns=["Complaint_ID","Complaint_Text","Language","Source","Resolution_Time_Hours"])
x.head()
```

```python
y = final.Resolution_Time_Hours
y.head()
```

```python
reg = linear_model.LinearRegression()
reg.fit(x,y)
```

```python
reg.predict(x)
```

```python
reg.score(x,y)
```

```python
import pickle
```

```python
with open("model_pickle",'wb') as file:
  pickle.dump(reg,file)
```
