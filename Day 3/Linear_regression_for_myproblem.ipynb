```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("dataset.csv")
df.head()
```

```python
priority_map = {"Low":1,"Medium":2,"High":3}
df["Priority"]=df["Priority"].map(priority_map)
df.head()
```

```python
plt.xlabel("Priority")
plt.ylabel("Resolution_Time")
plt.scatter(df["Priority"],df["Resolution_Time_Hours"],color="red",marker="*")
```

```python
priority = df["Priority"]
priority
```

```python
resolution_time = df["Resolution_Time_Hours"]
resolution_time
```

```python
reg = linear_model.LinearRegression()
reg.fit(priority.values.reshape(-1,1),resolution_time.values)
```

```python
reg.predict([[1]])
```

```python
reg.predict([[3]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.score(priority.values.reshape(-1,1),resolution_time.values)
```
